```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from openpyxl import load_workbook
from openpyxl.drawing.image import Image

# Step 1: Get the current working directory and create the 'image' subfolder
current_dir = os.getcwd()
image_dir = os.path.join(current_dir, "image")
if not os.path.exists(image_dir):
    os.makedirs(image_dir)

# Step 2: Generate sample data and save it to Excel
data = {'X': [1, 2, 3, 4, 5], 'Y': [2, 3, 5, 7, 11]}
df = pd.DataFrame(data)
excel_path = os.path.join(image_dir, "output.xlsx")
df.to_excel(excel_path, index=False, sheet_name="Data")  # Save data to Excel

# Step 3: Create and save the plot as an image in the 'image' subfolder
plt.figure(figsize=(6, 4))
plt.plot(df['X'], df['Y'], marker='o')
plt.title("Sample Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plot_path = os.path.join(image_dir, "plot.png")
plt.savefig(plot_path)

# Step 4: Insert the plot image into the Excel file
wb = load_workbook(excel_path)
ws = wb.create_sheet("Plot")
img = Image(plot_path)
ws.add_image(img, "A1")  # Insert image starting at cell A1
wb.save(excel_path)

print("Excel file with data and image saved at:", excel_path)
```
